**2.** Downloading videos from YouTube hosting and other options.
\
Загрузка видео с хостинга YouTube и другие опции.

```python
from ipywidgets import widgets
from IPython.display import display
import os

checkbox = widgets.Checkbox(
    value=False,
    description='Connect to Google Drive?\nПодключиться к Google Drive?',
    indent=False
)

def on_value_change(change):
    global output_dir
    if change['new']:
        from google.colab import drive
        drive.mount('/content/drive/')
        output_dir = '/content/drive/MyDrive/output_videos'
        if not os.path.exists(output_dir):
            os.makedirs(output_dir)
        print('Вы выбрали **диск** в качестве места для сохранения')
    else:
        output_dir = '/content/output_videos'
        if not os.path.exists(output_dir):
            os.makedirs(output_dir)
        print('Вы выбрали **временное хранилище файлов** в качестве места для хранения')

checkbox.observe(on_value_change, 'value')
display(checkbox)

checkbox = widgets.Checkbox(
    value=False,
    description='Конвертировать видео в аудио файл?',
    indent=False
)

def on_select_format(change):
  global options
  if change['new']:
    options = {
    'format': 'bestaudio/best',
    'extractaudio': True,
    'audioformat': 'mp3',
    'outtmpl': os.path.join(output_dir, '%(title)s.%(ext)s'),
    'restrictfilenames': True,
    'noplaylist': True,
    'nocheckcertificate': True,

    }

    print('Вы выбрали конвертацию видео в аудио')
  else:
    options = {'format': 'bestvideo+bestaudio/best', 'outtmpl': os.path.join(output_dir, '%(title)s.%(ext)s')}
    print('Вы выбрали установку видео')
checkbox.observe(on_select_format, 'value')
display(checkbox)
```

**3.** Downloading \
Скачивание

```python
import yt_dlp
import os
import shutil  # Импортируем модуль shutil

# Проверяем, определена ли переменная output_dir
try:
    output_dir
except NameError:
    output_dir = '/content/output_videos'  # Задаем значение по умолчанию
    print("Переменная output_dir не была определена. Используется временное хранилище.")

if os.path.exists(output_dir):
    answer = input('Желаете удалить старые сохранения?\nTrue - Да\nFalse - Нет\n')
    if answer == 'True':  # Проверяем строку напрямую
        shutil.rmtree(output_dir)  # Удаляем всю директорию
        print('Предыдущие сохранения были успешно удалены')
        os.makedirs(output_dir)  # Создаем директорию заново

url = input('Введите ссылку на ютуб видео или его id:')

if 'https://www.youtube.com/watch?v=' in url:
    pass  # Не нужно добавлять url заново, так как он уже правильный
elif 'https://youtu.be/' in url:
    pass
else:
    print('Проверьте вашу ссылку на корректность. Принимаются только ссылки с видеохостинга YouTube.com')
    exit()

ydl_opts = options
with yt_dlp.YoutubeDL(ydl_opts) as ydl:
    info_dict = ydl.extract_info(url, download=True)
    downloaded_file_path = ydl.prepare_filename(info_dict)
    print(f'{url} успешно загружен')

global zip_file_path
zip_file_path = f"{output_dir}/downloaded_videos.zip"
!zip -r {zip_file_path} {output_dir}/  # Убедитесь, что вы архивируете содержимое директории


```
