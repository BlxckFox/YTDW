**2.** Downloading videos from YouTube hosting and other options.
\
Загрузка видео с хостинга YouTube и другие опции.

```python
from ipywidgets import widgets
from IPython.display import display
import os

# Инициализируем глобальные переменные
output_dir = '/content/output_videos'  # Значение по умолчанию
options = {}

def initialize():
    global output_dir
    options = {}

checkbox = widgets.Checkbox(
    value=False,
    description='Connect to Google Drive?\nПодключиться к Google Drive?',
    indent=False
)

def on_value_change(change):
    global output_dir
    if change['new']:
        from google.colab import drive
        drive.mount('/content/drive/')
        output_dir = '/content/drive/MyDrive/output_videos'
        if not os.path.exists(output_dir):
            os.makedirs(output_dir)
        print('Вы выбрали **диск** в качестве места для сохранения')
    else:
        output_dir = '/content/output_videos'
        if not os.path.exists(output_dir):
            os.makedirs(output_dir)
        print('Вы выбрали **временное хранилище файлов** в качестве места для хранения')

checkbox.observe(on_value_change, 'value')
display(checkbox)

checkbox_audio = widgets.Checkbox(
    value=False,
    description='Конвертировать видео в аудио файл?',
    indent=False
)

def on_select_format(change):
    global options
    if change['new']:
        options = {
            'format': 'bestaudio/best',
            'extractaudio': True,
            'audioformat': 'mp3',
            'outtmpl': os.path.join(output_dir, '%(title)s.%(ext)s'),
            'restrictfilenames': True,
            'noplaylist': True,
            'nocheckcertificate': True,
        }
        output_template = os.path.join(output_dir, '%(title)s.webm')
        if 'webm' in output_template:
            options['outtmpl'] = os.path.join(output_dir, '%(title)s.mp3')
        print('Вы выбрали конвертацию видео в аудио')
    else:
        options = {'format': 'bestvideo+bestaudio/best', 'outtmpl': os.path.join(output_dir, '%(title)s.%(ext)s')}
        print('Вы выбрали установку видео')

checkbox_audio.observe(on_select_format, 'value')
display(checkbox_audio)

# Initializing
initialize()
```
